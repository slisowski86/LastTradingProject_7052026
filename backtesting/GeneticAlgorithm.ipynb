```python
# ----------------------------------------------------------------------
# 4. Utility functions for signals and constraints
# ----------------------------------------------------------------------
def get_signals(indicators):
    signals = []
    for ind in indicators:
        ind_name = ind.__class__.__name__
        for method_name in dir(ind):
            method = getattr(ind, method_name)
            if callable(method) and hasattr(method, '_signal_meta'):
                meta = method._signal_meta
                signals.append({
                    'indicator': ind,
                    'indicator_name': ind_name,
                    'method_name': method_name,
                    'category': ind.category,
                    'direction': meta['direction'],
                    'type': meta['type'],
                })
    return signals
```

```python
def build_signal_pool(indicators):
    pool = defaultdict(list)
    for s in get_signals(indicators):
        key = (s['category'], s['direction'], s['type'])
        pool[key].append((s['indicator'], s['method_name']))
    return pool
```

```python
def divergence_map_from_indicators(indicators):
    dm = defaultdict(list)
    for ind in indicators:
        base = ind.__class__.__name__
        for method_name in dir(ind):
            if 'divergence' in method_name.lower():
                dm[base].append(method_name)
    return dm
```

```python
def has_divergence_conflict(selected_methods, divergence_map):
    regular_bases = set()
    for ind, meth in selected_methods:
        base = ind.__class__.__name__
        is_div = any(meth == d for d in divergence_map.get(base, []))
        if not is_div:
            regular_bases.add(base)
    for ind, meth in selected_methods:
        base = ind.__class__.__name__
        is_div = any(meth == d for d in divergence_map.get(base, []))
        if is_div and base in regular_bases:
            return True
    return False
```

```python
# ----------------------------------------------------------------------
# 5. Random component & strategy generation (same as before)
# ----------------------------------------------------------------------
def random_component(comp_type, pool, used_indicators=None, divergence_map=None):
    if comp_type == 'entry_long':
        patterns_2, patterns_3 = PATTERNS_2_ENTRY_LONG, PATTERNS_3_ENTRY_LONG
    elif comp_type == 'exit_long':
        patterns_2, patterns_3 = PATTERNS_2_EXIT_LONG, PATTERNS_3_EXIT_LONG
    elif comp_type == 'entry_short':
        patterns_2, patterns_3 = PATTERNS_2_ENTRY_SHORT, PATTERNS_3_ENTRY_SHORT
    elif comp_type == 'exit_short':
        patterns_2, patterns_3 = PATTERNS_2_EXIT_SHORT, PATTERNS_3_EXIT_SHORT
    else:
        raise ValueError("Invalid component type")

    if random.random() < 0.5:
        pattern = random.choice(patterns_2)
    else:
        pattern = random.choice(patterns_3)

    used_local = set()
    result = []
    for spec in pattern:
        candidates = pool.get(spec, [])
        if not candidates:
            return None
        available = [(ind, name) for (ind, name) in candidates
                     if ind not in used_local and
                     (used_indicators is None or ind not in used_indicators)]
        if not available:
            return None
        ind, name = random.choice(available)
        result.append((ind, name))
        used_local.add(ind)
    if has_divergence_conflict(result, divergence_map):
        return None
    return result
```

```python
def random_strategy(indicators, pool, divergence_map, max_attempts=20):
    for _ in range(max_attempts):
        used_indicators = set()
        el = random_component('entry_long', pool, used_indicators, divergence_map)
        if el is None: continue
        used_indicators.update(ind for ind, _ in el)
        xl = random_component('exit_long', pool, used_indicators, divergence_map)
        if xl is None: continue
        used_indicators.update(ind for ind, _ in xl)
        es = random_component('entry_short', pool, used_indicators, divergence_map)
        if es is None: continue
        used_indicators.update(ind for ind, _ in es)
        xs = random_component('exit_short', pool, used_indicators, divergence_map)
        if xs is None: continue
        return {
            'entry_long': [f"{ind.__class__.__name__}.{name}" for ind, name in el],
            'exit_long':  [f"{ind.__class__.__name__}.{name}" for ind, name in xl],
            'entry_short':[f"{ind.__class__.__name__}.{name}" for ind, name in es],
            'exit_short': [f"{ind.__class__.__name__}.{name}" for ind, name in xs],
        }
    raise RuntimeError("Could not generate random strategy")
```

```python
@numba.njit(cache=True)
def fast_positions(entry_long, exit_long, entry_short, exit_short):
    n = len(entry_long)
    pos = np.zeros(n, dtype=np.int8)
    current = 0
    for i in range(n):
        if current == 0:
            if entry_long[i]:
                current = 1
            elif entry_short[i]:
                current = -1
        elif current == 1:
            if exit_long[i]:
                current = 0
        elif current == -1:
            if exit_short[i]:
                current = 0
        pos[i] = current
    return pos
```

```python
# ----------------------------------------------------------------------
# 6. Fast backtest components (Numba)
# ----------------------------------------------------------------------


def sharpe_and_return_from_positions(positions, returns):
    strat_returns = np.roll(positions, 1) * returns
    strat_returns[0] = 0
    strat_returns = strat_returns[~np.isnan(strat_returns)]
    if len(strat_returns) == 0:
        return -np.inf, 0
    total_return = np.prod(1 + strat_returns) - 1
    if np.std(strat_returns) == 0:
        return -np.inf, total_return
    sharpe = np.sqrt(252) * np.mean(strat_returns) / np.std(strat_returns)
    return sharpe, total_return
```

```python
# ----------------------------------------------------------------------
# 7. Pre‑compute raw signals (calls each indicator method once)
# ----------------------------------------------------------------------
def precompute_raw_signals(indicators, data):
    raw = {}
    meta = {}
    for ind in indicators:
        ind_name = ind.__class__.__name__
        for method_name in dir(ind):
            method = getattr(ind, method_name)
            if callable(method) and hasattr(method, '_signal_meta'):
                arr = method()
                if isinstance(arr, pd.Series):
                    arr = arr.values
                raw[f"{ind_name}.{method_name}"] = arr.astype(np.int8)
                meta[f"{ind_name}.{method_name}"] = {
                    'direction': method._signal_meta['direction'],
                    'type': method._signal_meta['type'],
                }
    return raw, meta

def binary_signal_from_raw(method_str, role, raw, meta):
    arr = raw[method_str]
    direction = meta[method_str]['direction']
    if role in ('entry_long', 'exit_short'):
        if direction == 'long':
            return arr == 1
        elif direction == 'both':
            return arr == 1
        else:
            return np.zeros(len(arr), dtype=bool)
    elif role in ('exit_long', 'entry_short'):
        if direction == 'short':
            return arr == -1
        elif direction == 'both':
            return arr == 1
        else:
            return np.zeros(len(arr), dtype=bool)
    else:
        return np.zeros(len(arr), dtype=bool)

def evaluate_strategy_metrics(strategy, raw, meta, returns, chunk_slice):
    """Return (sharpe, total_return, max_drawdown)."""
    # Build boolean arrays (same as before)
    el_bool = np.ones(len(returns), dtype=bool)
    for sig in strategy['entry_long']:
        if sig not in raw:
            return -np.inf, 0, 1.0
        el_bool &= binary_signal_from_raw(sig, 'entry_long', raw, meta)
    xl_bool = np.ones(len(returns), dtype=bool)
    for sig in strategy['exit_long']:
        if sig not in raw:
            return -np.inf, 0, 1.0
        xl_bool &= binary_signal_from_raw(sig, 'exit_long', raw, meta)
    es_bool = np.ones(len(returns), dtype=bool)
    for sig in strategy['entry_short']:
        if sig not in raw:
            return -np.inf, 0, 1.0
        es_bool &= binary_signal_from_raw(sig, 'entry_short', raw, meta)
    xs_bool = np.ones(len(returns), dtype=bool)
    for sig in strategy['exit_short']:
        if sig not in raw:
            return -np.inf, 0, 1.0
        xs_bool &= binary_signal_from_raw(sig, 'exit_short', raw, meta)

    el = el_bool[chunk_slice]
    xl = xl_bool[chunk_slice]
    es = es_bool[chunk_slice]
    xs = xs_bool[chunk_slice]
    rets = returns[chunk_slice]

    positions = fast_positions(el, xl, es, xs)
    strat_returns = np.roll(positions, 1) * rets
    strat_returns[0] = 0
    strat_returns = strat_returns[~np.isnan(strat_returns)]
    if len(strat_returns) == 0:
        return -np.inf, 0, 1.0

    total_return = np.prod(1 + strat_returns) - 1
    if np.std(strat_returns) == 0:
        sharpe = -np.inf
    else:
        sharpe = np.sqrt(252) * np.mean(strat_returns) / np.std(strat_returns)

    # Compute max drawdown
    cum_returns = np.cumprod(1 + strat_returns)
    running_max = np.maximum.accumulate(cum_returns)
    drawdown = (cum_returns - running_max) / running_max
    max_drawdown = np.min(drawdown)  # negative number, e.g., -0.15

    return sharpe, total_return, max_drawdown
```

```python
# ----------------------------------------------------------------------
# 8. Walk‑forward testing (same as before)
# ----------------------------------------------------------------------
def walk_forward_test(strategy, raw, meta, returns, chunks, min_sharpe_list, min_return_list):
    for i, chunk in enumerate(chunks):
        sharpe, ret = evaluate_strategy_on_chunk(strategy, raw, meta, returns, chunk)
        if sharpe < min_sharpe_list[i]:# or ret < min_return_list[i]:
            return False, None, None
    last_sharpe, last_ret = evaluate_strategy_on_chunk(strategy, raw, meta, returns, chunks[-1])
    return True, last_sharpe, last_ret
```

```python
def evaluate_strategy_on_chunk(strategy, raw, meta, returns, chunk_slice):
    sharpe, _, _ = evaluate_strategy_metrics(strategy, raw, meta, returns, chunk_slice)
    return sharpe, None   # second value unused
```

```python
# ----------------------------------------------------------------------
# 9. Genetic Algorithm core
# ----------------------------------------------------------------------
class StrategyGA:
    def __init__(self, indicators, pool, divergence_map, raw, meta, returns, pre_screen_chunk,
                 pop_size=100, generations=30, mutation_rate=0.2, crossover_rate=0.8,
                 elite_size=5, max_global_usage=2):
        self.indicators = indicators
        self.pool = pool
        self.divergence_map = divergence_map
        self.raw = raw
        self.meta = meta
        self.returns = returns
        self.pre_screen_chunk = pre_screen_chunk
        self.pop_size = pop_size
        self.generations = generations
        self.mutation_rate = mutation_rate
        self.crossover_rate = crossover_rate
        self.elite_size = elite_size
        self.max_global_usage = max_global_usage

    def fitness(self, strategy):
        sharpe, _ = evaluate_strategy_on_chunk(strategy, self.raw, self.meta, self.returns, self.pre_screen_chunk)
        return sharpe

    def create_individual(self):
        return random_strategy(self.indicators, self.pool, self.divergence_map)

    def initial_population(self):
        pop = []
        while len(pop) < self.pop_size:
            try:
                ind = self.create_individual()
                pop.append(ind)
            except RuntimeError:
                continue
        return pop

    def tournament_selection(self, population, fitness_scores, k=3):
        selected = []
        for _ in range(len(population)):
            idx = random.sample(range(len(population)), k)
            best = idx[np.argmax([fitness_scores[i] for i in idx])]
            selected.append(population[best])
        return selected

    def crossover(self, parent1, parent2):
        if random.random() > self.crossover_rate:
            return parent1.copy(), parent2.copy()
        # single point crossover over components
        point = random.randint(0, 3)
        keys = ['entry_long', 'exit_long', 'entry_short', 'exit_short']
        child1, child2 = {}, {}
        for i, k in enumerate(keys):
            if i <= point:
                child1[k] = parent1[k][:]
                child2[k] = parent2[k][:]
            else:
                child1[k] = parent2[k][:]
                child2[k] = parent1[k][:]
        # Validate children (global usage, divergence, distinct indicators)
        if self.is_valid_strategy(child1) and self.is_valid_strategy(child2):
            return child1, child2
        return parent1.copy(), parent2.copy()

    def mutate(self, strategy):
        if random.random() > self.mutation_rate:
            return strategy
        # mutate a random component
        keys = ['entry_long', 'exit_long', 'entry_short', 'exit_short']
        comp = random.choice(keys)
        # compute used indicators from other three components
        used_inds = set()
        for k in keys:
            if k != comp:
                for sig in strategy[k]:
                    ind_name = sig.split('.')[0]
                    # find indicator object by name
                    for ind in self.indicators:
                        if ind.__class__.__name__ == ind_name:
                            used_inds.add(ind)
                            break
        # generate new component
        new_comp = random_component(comp, self.pool, used_inds, self.divergence_map)
        if new_comp is not None:
            strategy[comp] = [f"{ind.__class__.__name__}.{name}" for ind, name in new_comp]
        return strategy

    def is_valid_strategy(self, strategy):
        # check per‑component distinct indicators and divergence conflict
        # also global usage <= max_global_usage
        usage = Counter()
        for comp in ['entry_long', 'exit_long', 'entry_short', 'exit_short']:
            comp_inds = []
            for sig in strategy[comp]:
                ind_name = sig.split('.')[0]
                # find indicator object
                for ind in self.indicators:
                    if ind.__class__.__name__ == ind_name:
                        comp_inds.append(ind)
                        usage[ind] += 1
                        break
            # check duplicates within component
            if len(set(comp_inds)) != len(comp_inds):
                return False
            # check divergence conflict (need method names)
            # build list of (ind, method_name) for this component
            comp_methods = []
            for sig in strategy[comp]:
                ind_name, meth = sig.split('.')
                for ind in self.indicators:
                    if ind.__class__.__name__ == ind_name:
                        comp_methods.append((ind, meth))
                        break
            if has_divergence_conflict(comp_methods, self.divergence_map):
                return False
        # global usage limit
        if any(cnt > self.max_global_usage for cnt in usage.values()):
            return False
        return True

    def run(self, verbose=True):
        population = self.initial_population()
        best_fitness = -np.inf
        best_strategy = None
        best_metrics = None   # (sharpe, return, drawdown)
    
        for gen in tqdm(range(self.generations), desc="GA generations"):
            fitness_scores = [self.fitness(s) for s in population]
            gen_best_idx = np.argmax(fitness_scores)
            gen_best_fitness = fitness_scores[gen_best_idx]
    
            if gen_best_fitness > best_fitness:
                best_fitness = gen_best_fitness
                best_strategy = population[gen_best_idx].copy()
                # Compute full metrics using the pre‑screen chunk
                sharpe, ret, dd = evaluate_strategy_metrics(
                    best_strategy, self.raw, self.meta, self.returns, self.pre_screen_chunk
                )
                best_metrics = (sharpe, ret, dd)
    
                if verbose:
                    print(f"\n{'='*60}")
                    print(f"Gen {gen} – New best strategy found")
                    print(f"  Sharpe:       {sharpe:.4f}")
                    print(f"  Total Return: {ret:.4f}")
                    print(f"  Max Drawdown: {dd:.4f}")
                    print(f"  Entry Long:   {best_strategy['entry_long']}")
                    print(f"  Exit Long:    {best_strategy['exit_long']}")
                    print(f"  Entry Short:  {best_strategy['entry_short']}")
                    print(f"  Exit Short:   {best_strategy['exit_short']}")
                    print(f"{'='*60}\n")
    
            # --- selection, crossover, mutation, elitism (unchanged) ---
            selected = self.tournament_selection(population, fitness_scores)
            next_pop = []
            # elitism
            elite_idx = np.argsort(fitness_scores)[-self.elite_size:]
            for idx in elite_idx:
                next_pop.append(population[idx].copy())
            while len(next_pop) < self.pop_size:
                p1 = random.choice(selected)
                p2 = random.choice(selected)
                c1, c2 = self.crossover(p1, p2)
                c1 = self.mutate(c1)
                c2 = self.mutate(c2)
                next_pop.append(c1)
                if len(next_pop) < self.pop_size:
                    next_pop.append(c2)
            population = next_pop
    
        self.population = population
        return best_strategy, best_fitness, best_metrics
```

```python
def ga_pipeline(indicators, data,
                pop_size=100, generations=30,
                mutation_rate=0.2, crossover_rate=0.8,
                pre_screen_bars=5000,
                min_sharpe=0.5,
                min_return=0.01,
                max_drawdown=-0.20,
                top_n_candidates=50,
                output_file="ga_best_strategies.jsonl",
                verbose=True):
    """
    Genetic Algorithm pipeline for strategy discovery.
    Returns a list of dictionaries containing strategies that pass all filters.
    """
    # 1. Pre‑compute raw signals and returns
    raw, meta = precompute_raw_signals(indicators, data)
    returns = data['Close'].pct_change().fillna(0).values
    pool = build_signal_pool(indicators)
    div_map = divergence_map_from_indicators(indicators)

    pre_screen_chunk = slice(0, pre_screen_bars)

    # 2. Run GA
    ga = StrategyGA(indicators, pool, div_map, raw, meta, returns, pre_screen_chunk,
                    pop_size=pop_size, generations=generations,
                    mutation_rate=mutation_rate, crossover_rate=crossover_rate)
    best_strategy, best_sharpe, best_metrics = ga.run(verbose=verbose)
    best_sharpe_full, best_return, best_dd = best_metrics

    print("\n" + "="*60)
    print("GA FINISHED – Best overall strategy (on pre‑screen chunk)")
    print(f"  Sharpe:       {best_sharpe_full:.4f}")
    print(f"  Total Return: {best_return:.4f}")
    print(f"  Max Drawdown: {best_dd:.4f}")
    print(f"  Entry Long:   {best_strategy['entry_long']}")
    print(f"  Exit Long:    {best_strategy['exit_long']}")
    print(f"  Entry Short:  {best_strategy['entry_short']}")
    print(f"  Exit Short:   {best_strategy['exit_short']}")
    print("="*60)

    # 3. Evaluate final population (all individuals) and filter
    final_population = ga.population
    passing = []
    for strat in tqdm(final_population, desc="Evaluating final candidates"):
        sharpe, ret, dd = evaluate_strategy_metrics(strat, raw, meta, returns, pre_screen_chunk)
        if sharpe >= min_sharpe and ret >= min_return and dd >= max_drawdown:
            passing.append({
                'strategy': strat,
                'sharpe': sharpe,
                'total_return': ret,
                'max_drawdown': dd
            })

    # 4. Sort by Sharpe and keep top N
    passing.sort(key=lambda x: x['sharpe'], reverse=True)
    passing = passing[:top_n_candidates]

    # 5. Save to file with informative prints
    if passing:
        print(f"\n✅ Saving {len(passing)} strategies that passed filters:")
        print(f"   - Minimum Sharpe: {min_sharpe}")
        print(f"   - Minimum Return: {min_return}")
        print(f"   - Maximum Drawdown: {max_drawdown}")
        print(f"   - File: {output_file}\n")

        # Show top 3
        print("🏆 Top 3 strategies by Sharpe:")
        for i, rec in enumerate(passing[:3], 1):
            print(f"  {i}. Sharpe: {rec['sharpe']:.4f}, Return: {rec['total_return']:.4f}, DD: {rec['max_drawdown']:.4f}")
            print(f"     Entry Long: {rec['strategy']['entry_long']}")
            print(f"     Exit Long:  {rec['strategy']['exit_long']}")
            if i < 3:
                print()

        with open(output_file, 'w') as f:
            for rec in passing:
                f.write(json.dumps(rec) + '\n')
        print(f"\n💾 Strategies saved to {output_file}")
    else:
        print(f"\n⚠️ No strategies passed the filters. Try relaxing min_sharpe, min_return, or max_drawdown.")

    return passing
```
